```python
from getpass import getpass
```

```python
api_key = getpass("APIキーを入力してください")
```

```python
search_word = input("駅名を入力してください")
```

```python
url = "https://maps.googleapis.com/maps/api/geocode/json"
```

```python
response = requests.get(
    url,
    params = {
        'key': api_key,
        'address': search_word
    }
)
```

```python
json_data = response.json()
```

```python
pprint.pprint(json_data)
```

```python
latitude = 35.7185858
longitude = 139.7958849
radius = 500
language = "ja"

# Google Places APIのURL
url2 = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

response = requests.get(url2, params={
    'key': api_key,
    'location': f'{latitude},{longitude}',
    'radius': radius,
    'type': 'restaurant',
    'language': language
})

json_data = response.json()
pprint.pprint(json_data)
results = json_data.get('results', [])
```

```python
for restaurant in results:
    name = restaurant.get('name')
    rating = restaurant.get('rating')
    address = restaurant.get('vicinity')
    print(f"レストラン名: {name}, 評価: {rating}, 住所: {address}")
```
